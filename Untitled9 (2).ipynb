```python
import numpy as np  

import matplotlib.pyplot as mtp    

import pandas as pd    
```

```python
data=pd.read_csv('/content/heart disease classification dataset.csv')
data.head(5)

```

```python
del data['Unnamed: 0']
```

```python
data.drop(data.index[range(20, 303)], axis=0, inplace=True)

```

```python
from sklearn.cluster import KMeans  
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', random_state= 42)  
```

```python
y_predict= kmeans.fit_predict(data) 
```

```python
data=np.array(data)
```

```python
mtp.scatter(data[y_predict == 0, 0], data[y_predict == 0, 1], s = 100, c = 'blue', label = 'Non aginial pain') #for first cluster  

mtp.scatter(data[y_predict == 1, 0], data[y_predict == 1, 1], s = 100, c = 'green', label = 'Typical Aginia') #for second cluster  

mtp.scatter(data[y_predict== 2, 0], data[y_predict == 2, 1], s = 100, c = 'red', label = 'Asysmptotic') #for third cluster

mtp.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')

mtp.title('clusters of Types of chest pain')  

mtp.xlabel('Age')  

mtp.ylabel('Type of chest pain')  

mtp.legend()  

mtp.show()  
```
